```python
import numpy as np

import matplotlib.pyplot as plt
%matplotlib inline
```

```python
AU = 1.49598e+13 # Average distance from the Earth to the Sun in cm
G = 6.67259e-08 # universal gravitational constant in cgs units 
msun = 1.9891e33 # mass of the Sun in grams
yr =  365.25 * 24 * 60 * 60 # 1 year in seconds
kpc = 3.08568e+21 # kiloparsec in cm
```

### <font color='black'> $N$-body integration

Implementing a function that carries out integrations of $N$-body dynamics of $N$ particles using the Kick-Drift (KD) update scheme. 
    
This function takes as input the name of a function that computes accelerations for every particle using coordinates. 

```python
def read_solsys_bodies(plist = []):
    index_dict = {'Sun': 9, 'Mercury': 0, 'Venus': 1, 'Earth': 2, 'Mars': 3, 
                  'Jupiter': 4, 'Saturn': 5, 'Uranus': 6, 'Neptune': 7, 'Pluto': 8}
    
    data_file = 'https://files/~andrey/classes/a211/data/planets.dat'
    name = np.loadtxt(data_file, usecols=[0], unpack=True, dtype=str)

    mp, xp, yp, zp, vxp, vyp, vzp = np.loadtxt(data_file, usecols=(1,2,3,4,5,6,7), unpack=True)

    AU = 1.49598e+13 
    msun = 1.9891e33 
    aupd = AU / (24*3600.)
    
    out_dict = {}
    for name in plist:
        assert(name in index_dict)
        ip = index_dict[name]
        out_dict[name] = {'mass': mp[ip]*msun, 'x': xp[ip]*AU, 'y': yp[ip]*AU, 'z': zp[ip]*AU,
                          'vx':  vxp[ip]*aupd, 'vy': vyp[ip]*aupd, 'vz': vzp[ip]*aupd}

    return out_dict
```

```python
def acc_func(x, mp, G=6.67259e-08):
    N = x.shape[0]
    acc = np.zeros_like(x)
    for i in range(N):
        for j in range(N):
            if i != j:
                r = x[j] - x[i]
                r_mag = np.sqrt(r[0]**2 + r[1]**2 + r[2]**2)
                if r_mag > 0:
                    acc[i] += G * mp[j] * r / r_mag**3
    return acc
    
def integration(x, v, mp, dt=None, nsteps=None, acc_func=None):
    N = x.shape[0]
    tt = np.zeros(nsteps)
    xt = np.zeros((nsteps, N, 3))
    vt = np.zeros((nsteps, N, 3))

    xt[0] = x
    vt[0] = v

    for n in range(nsteps-1):
        a = acc_func(xt[n], mp)
        vt[n+1] = vt[n] + a * dt
        xt[n+1] = xt[n] + vt[n+1] * dt 

        tt[n+1] = tt[n] + dt

    return tt, xt, vt

#Testing function by integrating two body system with the Sun and Earth
out = read_solsys_bodies(plist = ['Sun', 'Earth'])

m_e = out['Earth']['mass']
m_s = out['Sun']['mass']
x, y, z = out['Earth']['x'], out['Earth']['y'], out['Earth']['z']
vx, vy, vz = out['Earth']['vx'], out['Earth']['vy'], out['Earth']['vz']
xsun, ysun, zsun = out['Sun']['x'], out['Sun']['y'], out['Sun']['z']
vx_sun, vy_sun, vz_sun = out['Sun']['vx'], out['Sun']['vy'], out['Sun']['vz']

initial_pos = np.array([[xsun, ysun, zsun], [x, y, z]])
initial_vel = np.array([[vx_sun, vy_sun, vz_sun], [vx, vy, vz]])
masses = np.array([m_s, m_e])

nsteps = 100000
dt = 10 * yr / nsteps

tt, xt, vt = integration(initial_pos, initial_vel, masses, dt=dt, nsteps=nsteps, acc_func=acc_func)

# plotting the distance of the Earth from the Sun, calculating eccentricity of Earth's orbit

distances = np.sqrt(np.sum((xt[:, 1] - xt[:, 0])**2, axis=1))
r_min = np.min(distances)
r_max = np.max(distances)
e = (1-(r_min/r_max))/(1+(r_min/r_max))

plt.figure(figsize=(6, 6))
plt.plot(xt[:, 1, 0], xt[:, 1, 1])
plt.xlabel('x position')
plt.ylabel('y position')
plt.title('Earth orbit')
plt.grid(True)
plt.show()

plt.figure(figsize=(6, 6))
plt.plot(tt, distances, label='Earth-Sun distance')
plt.xlabel('Time (years)')
plt.ylabel('Distance (AU)')
plt.title('Earth-Sun Distance Over Time')
plt.grid(True)
plt.legend()
plt.show()

print(f"r_min/r_max: {r_min/r_max:.6f}")
print(f"Eccentricity: {e:.6f}")
```

### <font color='black'>Which planets of the Solar system would be detectable if we observed them around another star?</font>

Exoplanets can be detected by the gravitational pull they exert on the star they orbit. If the planet is sufficiently massive, the motion of the star can be detected using spectral lines in its spectrum and the Doppler Shift. If more than one planet exerts a significant pull on the star, these multiple pulls will be reflected in an oscillatory change of velocity over time where multiple oscillation frequencies are present. This can be used to map out cyclical motion of the star.

With current technology we can only detect velocity changes of about 1 meter per second and larger.

```python
def sun_vel(tt, xt, vt):
    vx = vt[:, 0, 0]
    vx *= 1e-2

    vx_mean = np.mean(vx)
    osc = vx - vx_mean

    tt /= yr

    plt.figure(figsize=(4, 4))
    plt.plot(tt, osc, label='Vx oscillations')
    plt.xlabel('Time (years)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Sun Vx Velocity Oscillations - Earth')
    plt.grid(True)
    plt.show()

    max_oscillation = np.max(np.abs(osc))
    print(f"Maximum Vx oscillation amplitude: {max_oscillation:.3f} m/s")
    if max_oscillation > 1.0:
        print("The Earth is detectable.")
    else:
        print("The Earth is not detectable, because the oscillation amplitude is less than the velocity measurement accuracy of 1 m/s.")

# Plotting x component of the Sun's velocity (in m/s) as a function of time (in years) using 
# Sun's trajectory from previous Earth-Sun system 

# Calculating the evolution of the Earth-Sun system for at least 10 orbital periods of the Earth
# Plotting the velocity component as a function of time
nsteps = 100000
dt = 10 * yr / nsteps
tt, xt, vt = integration(initial_pos, initial_vel, masses, dt=dt, nsteps=nsteps, acc_func=acc_func)

sun_vel(tt, xt, vt)

#Plotting x components of Sun's velocity with Mercury, Venus, Mars, Jupiter, Saturn, Neptune, Uranus
def planet_sim(m_sun, mplan, initial_pos, initial_vel, orbits, dt, acc_func, planet_name, G = 6.67259e-08):
    a = np.sqrt(np.sum((initial_pos[1] - initial_pos[0])**2))
    period = (2 * np.pi)/(np.sqrt(G*(m_sun + mplan))) * a**(3/2)
    
    total_time = orbits * period
    nsteps = int(total_time/dt)
      
    masses = np.array([m_sun, mplan])
    tt, xt, vt = integration(initial_pos, initial_vel, masses, dt=dt, nsteps=nsteps, acc_func=acc_func)

    vx = vt[:, 0, 0]
    vx *= 1e-2

    vx_mean = np.mean(vx)
    osc = vx - vx_mean

    tt /= yr

    plt.figure(figsize=(4, 4))
    plt.plot(tt, vx - vx_mean, label='Sun X-Velocity')
    plt.xlabel('Time (years)')
    plt.ylabel('Velocity (cm/s)')
    plt.title(f'Sun Motion - {planet_name}')
    plt.grid(True)
    plt.show()

    max_oscillation = np.max(np.abs(osc))
    print(f"Maximum Vx oscillation amplitude: {max_oscillation:.3f} m/s")
    if max_oscillation > 1.0:
        print(f'{planet_name} is detectable.')
    else:
        print(f'{planet_name} is not detectable, because the oscillation amplitude is less than the velocity measurement accuracy of 1 m/s.')

def sim_loop(sun_pos, sun_vel, plan_data, m_sun, dt, acc_func):
    for planet_name, plan_data in plan_data.items():
        initial_pos = np.array([[sun_pos[0], sun_pos[1], sun_pos[2]], [plan_data['x'], plan_data['y'], plan_data['z']]])
        initial_vel = np.array([[sun_vel[0], sun_vel[1], sun_vel[2]], [plan_data['vx'], plan_data['vy'], plan_data['vz']]])
        planet_sim(m_sun, plan_data['mass'], initial_pos, initial_vel, orbits=10, dt=dt, acc_func=acc_func, planet_name=planet_name, G = 6.67259e-08)

out = read_solsys_bodies(plist=['Sun'])
m_sun = out['Sun']['mass']
xsun, ysun, zsun = out['Sun']['x'], out['Sun']['y'], out['Sun']['z']
vx_sun, vy_sun, vz_sun = out['Sun']['vx'], out['Sun']['vy'], out['Sun']['vz']

sun_pos = np.array([xsun, ysun, zsun])
sun_vel = np.array([vx_sun, vy_sun, vz_sun])

plan_data = read_solsys_bodies(plist=['Mercury', 'Venus', 'Mars', 'Jupiter', 'Saturn', 'Uranus', 'Neptune'])
sim_loop(sun_pos, sun_vel, plan_data, m_sun, dt, acc_func)
```

```python
#calculating evolution of the Sun system with all of planets that are detectable simultaneously 
# Plotting the component of Sun's velocity as a function of time for at least 100 years 
def sun_vel_detect(tt, xt, vt):
    vx = vt[:, 0, 0]
    vx *= 1e-2

    vx_mean = np.mean(vx)
    osc = vx - vx_mean

    tt /= yr

    plt.figure(figsize=(4, 4))
    plt.plot(tt, osc, label='Vx oscillations')
    plt.xlabel('Time (years)')
    plt.ylabel('Velocity (m/s)')
    plt.title('Sun Vx Velocity Oscillations - Combined Planets')
    plt.grid(True)
    plt.show()

#using integration function from before:
out = read_solsys_bodies(plist=['Sun', 'Jupiter', 'Saturn'])
m_sun = out['Sun']['mass']
xsun, ysun, zsun = out['Sun']['x'], out['Sun']['y'], out['Sun']['z']
vx_sun, vy_sun, vz_sun = out['Sun']['vx'], out['Sun']['vy'], out['Sun']['vz']

xjupiter, yjupiter, zjupiter = out['Jupiter']['x'], out['Jupiter']['y'], out['Jupiter']['z']
vx_jupiter, vy_jupiter, vz_jupiter = out['Jupiter']['vx'], out['Jupiter']['vy'], out['Jupiter']['vz']
xsaturn, ysaturn, zsaturn = out['Saturn']['x'], out['Saturn']['y'], out['Saturn']['z']
vx_saturn, vy_saturn, vz_saturn = out['Saturn']['vx'], out['Saturn']['vy'], out['Saturn']['vz']
m_jupiter = out['Jupiter']['mass']
m_saturn = out['Saturn']['mass']

pos = np.array([[xsun, ysun, zsun], [xjupiter, yjupiter, zjupiter], [xsaturn, ysaturn, zsaturn]])
vel = np.array([[vx_sun, vy_sun, vz_sun], [vx_jupiter, vy_jupiter, vz_jupiter], [vx_saturn, vy_saturn, vz_saturn]])
masses2 = np.array([m_sun, m_jupiter, m_saturn])

nsteps = 100000
dt = 100 * yr / nsteps
tt2, xt2, vt2 = integration(pos, vel, masses2, dt=dt, nsteps=nsteps, acc_func=acc_func)

sun_vel_detect(tt2, xt2, vt2)
```

Discussion: 
There is a pretty clear visual difference when looking at the plots created with a 2-body system vs. an N>2 body system. The plots with just a single planet look even in amplitude; in contrast, the multiple-planet plot has peaks and valleys of varying heights. This makes sense, because the planets are all in different places and orbits, and thus affect the Sun's motion non-uniformly.
