## <font color='black'>Implementing numerical integration with error control</font>

Implementing a function <tt>integrate</tt> that can numerically estimate integral for an input function over an interval $[a,b]$ with error control to stop iteration. The estimate uses the Romberg integration method to ensure that the estimate returned by the function has an estimated uncertainty smaller than the specified thresholds ${\rm atol}$ and ${\rm rtol}$.

Iterations stop when:

$$\vert R_{m}(h)-R_{m-1}(h/2)\vert<{\rm atol + rtol}\,\vert R_m(h)\vert,$$

where ${\rm atol}$ and ${\rm rtol}$ are the absolute and relative (fractional) error tolerances.

When ${\rm atol>rtol}\,\vert R_m(h)\vert$ the iteration will be stopped when the absolute error threshold is satisfied. When ${\rm atol<rtol}\,\vert R_m(h)\vert$, the iteration will be stopped when the relative error threshold is satisfied. 

```python
def trapzd_h(func, a, b, h, *args):
    """
    given function implementing trapezoidal integration scheme using
    numpy vector operations
    """
    assert(h > 0); assert(a < b)
    ntrap = (b-a) / h
    # assert that [a,b] is divided by h into integer number of intervals
    if ntrap - int(ntrap) != 0.:
        raise ValueError(f'interval [{a:.3e}, {b:.3e}] must be exactly divisible by h = {h:.3e}')
    # if h = b - a
    if abs(h - (b-a)) + 1. == 1.:
        return 0.5*(func(a, *args) + func(b, *args)) * (b-a)
    else: 
        hstep = (b - a) / ntrap
        # np.arange creates an array of values from 0 to nstep-1
        xd = a + np.arange(1,ntrap) * hstep 
        return (0.5*(func(a, *args) + func(b, *args)) + np.sum(func(xd, *args))) * hstep
```

```python
def romberg_recursive(func, a, b, h, *args, m=0):
    """
    Recursive implementation of Romberg integration of 
    the integral int_a^b f(x)dx with initial stepsize h
    up to the Romberg iteration m
    """
    if m == 0: 
        return trapzd_h(func, a, b, h, *args)
    else:
        fact = 4**m 
        # R_m-1(h/2)
        r_mm1_hhalf = romberg_recursive(func, a, b, h/2, *args, m=m-1)
        # R_m-1(h)
        r_mm1_h = romberg_recursive(func, a, b, h, *args, m=m-1)
        # compute R_m(h) using recursion relation 
        rm_h = (fact*r_mm1_hhalf - r_mm1_h) / (fact - 1)
        return rm_h
```

```python
def integrate(func, a, b, *args, atol=1e-10, rtol=1e-6, m_max=10):
    h = b - a
    for m in range (0, m_max + 1): 
        if m == 0: 
            prev = trapzd_h(func, a, b, h, *args)
        else:
            current = romberg_recursive(func, a, b, h, *args, m=m)
            error = abs(current - prev)
            if error < atol + (rtol * abs(current)): 
                return current, error
            prev = current

# Testing function by integrating e^x
def func(x):
    return np.exp(x)

exact_value = np.e - 1
rtols = [1e-3, 1e-8, 1e-12]

print('Varying rtol Values:')

for rtol in rtols: 
    rm_h, error = integrate(func, 0, 1, atol=1e-16, rtol=rtol, m_max=10)
    frac_error = abs(rm_h - exact_value) / abs(exact_value)
    print(f"Estimate: {rm_h}\nError: {error}\nrtol: {rtol}\nFractional Error: {frac_error}\nFractional Error < rtol\n")

print("\n")

# Repeat the test with varying atol 
exact_value = np.e - 1
atols = [1e-3, 1e-5, 1e-7]

print('Varying atol Values:')

for atol in atols: 
    rm_h, error = integrate(func, 0, 1, atol=atol, rtol=1e-16, m_max=10)
    frac_error = abs(rm_h - exact_value) / abs(exact_value)
    print(f"Estimate: {rm_h}\nError: {error}\natol: {atol}\nFractional Error: {frac_error}\nFractional Error < atol\n")
```

## <font color='black'>Using function to compute luminosity distance $d_L$</font>

Using previous integration function below in this line of given function d_l: 

            dc = integrate(dl_func, 0, z, *args, atol=atol, rtol=rtol)[0]

To compute the cosmological distance $d_L$ for a redshift $z$. 

```python
def d_L_no_de(z, H0, Om0, clight = 2.99792458e5):
    assert(Om0 >=0)
    q0 = 0.5 * Om0
    q0z = q0 * z
    return clight * z/H0 * (1. + (z-q0z) / (1. + q0z + np.sqrt(1. + 2.*q0z)))  
```

```python
def dl_func(z, H0, Om0, OmL, Omk):
    z1 = 1.0 + z; z12 = z1 * z1
    return 1.0 / np.sqrt(z12*(Om0*z1 + Omk) + OmL)
```

```python
clight = 2.99792458e5 # speed of light in km/s

def d_l(z, H0, Om0, OmL, atol=1e-16, rtol=1.e-8):
    """
    given function computing luminosity distance from z=0 to z 
    using helper routine for comoving distance _dc
    """
    # if OmL == 0
    if 1.0+OmL == 1.0: 
        return d_L_no_de(z, H0, Om0) # if OmL=0, use simple formula without integration
    else: # otherwise use numerical integration to compute distance 
        dH = clight / H0 
        Omk = 1.0 - Om0 - OmL
        args = [H0, Om0, OmL, Omk]
        dc = integrate(dl_func, 0, z, *args, atol=atol, rtol=rtol)[0]
        if 1.0 + Omk == 1.0:
            return dH * dc * (1.0 + z)
        else:
            sqrOmk = np.sqrt(np.abs(Omk))
            if Omk > 0.:
                return dH * np.sinh(dc*sqrOmk) * (1.0 + z) / sqrOmk 
            else:
                return dH * np.sin(dc*sqrOmk) * (1.0 + z) / sqrOmk
```

```python
# d_L function from Astropy for testing
from astropy.cosmology import LambdaCDM
import astropy.units as u

def d_l_astropy(z, H0, om0, oml):
    cosmo = LambdaCDM(H0=H0, Om0=om0, Ode0=oml)
    return cosmo.luminosity_distance(z=z) / u.Mpc
```

```python
#Testing function by comparing output to AstroPy estimates with several pairs of Om0 and OmL
omegas = [[0.3, 0.7], [0.1, 0.9], [0.8, 0.1]]
H0 = 70

print('z = 1:\n')
for Om0, OmL in omegas:
    dl_result = d_l(1, H0, Om0, OmL, atol=1e-16, rtol=1.e-8)
    dl_astropy = d_l_astropy(1, H0, Om0, OmL)
    
    print(f'd_L: {dl_result}\nAstropy d_L: {dl_astropy}\n')

print('\n')

print('z = 15:\n')
for Om0, OmL in omegas:
    dl_result = d_l(15, H0, Om0, OmL, atol=1e-16, rtol=1.e-8)
    dl_astropy = d_l_astropy(15, H0, Om0, OmL)
    
    print(f'd_L: {dl_result}\nAstropy d_L: {dl_astropy}\n')
```

```python
# Determining accuracy experimentally using <tt>d_l</tt> function
rtols = [1e-6, 1e-10, 1e-13, 1e-15]
Om0 = 0.3
OmL = 0.7

for rtol in rtols: 
    dl_result = d_l(1, H0, Om0, OmL, atol=1e-16, rtol=rtol)
    dl_astropy = d_l_astropy(1, H0, Om0, OmL)
    frac_diff = abs(dl_astropy - dl_result) / dl_astropy

    print(f'rtol = {rtol}:\nd_L: {dl_result}\nAstropy d_L: {dl_astropy}\nFractional Difference: {frac_diff}\n')
    
```

Discussion:
The fractional difference between my d_L function and astropy's d_L function stabilizes once rtol reaches 1e-13, meaning that the astropy function likely computes d_L with a precision of 1e-13. 

## <font color='black'>Using calculation to compute distance modulus of supernovae Type Ia</font>
    
Here I assume that supernovae are perfect standard candles, which means that $M$ has a single value for all of them. This means that we should be able to predict what apparent magnitudes of supernovae should be at different redshifts: 
    
$$m_{\rm pred} = M + 5\log_{10}d_L(z,H_0,\Omega_{\rm m0},\Omega_\Lambda) + 25.$$

Distance modulus:
    
$$\mu = m - M = 5\log_{10}d_L(z,H_0,\Omega_{\rm m0},\Omega_\Lambda) + 25.$$

Array <tt>zsn</tt> contains redshifts of 740 supernovae, while arrays <tt>msn</tt>, $m$, and <tt>emsn</tt> contain measured apparent magnitudes and their uncertainties. In this code I plot supernovae data as a scatter of points in the $m-z$ plane and plot $m_{\rm pred}(z)$ for a grid of $z$ values in the range $z\in [0,2]$ as a line.
    

```python
def read_jla_data(sn_list_name = None):
    zsn, msn, emsn = np.loadtxt(sn_list_name, usecols=(1, 4, 5),  unpack=True)

    return zsn, msn, emsn
```

```python
zsn, msn, emsn = read_jla_data(sn_list_name = 'https://files/~andrey/classes/a211/data/jla_lcparams.txt')
nsn = np.size(zsn)
print("read sample of %d supernovae..."%(nsn))
```

```python
# plot data and predicted magnitudes
Om0=0.5
OmL=0.5
H0=70
M=-19

z_grid = np.linspace(0.1, 2, 100)

dl = []
m_predicted = []
for z_value in z_grid:
    d_L = d_l(z_value, H0, Om0, OmL, atol=1e-16, rtol=1.e-8)
    m_pred = M + 5*np.log10(d_L) + 25
    dl.append(d_L)
    m_predicted.append(m_pred)

plt.figure(figsize=(7, 7))
plt.scatter(zsn, msn, marker='.', color='indigo', s=20, label='Supernovae Magnitude Data')
plt.plot(z_grid, m_predicted, color='red', lw=1, label='Predicted Magnitude')
plt.xlabel('z (Redshift)', fontsize=14)
plt.ylabel('Apparent Magnitude', fontsize=14)
plt.legend(loc='best', frameon=False)
plt.show()
```

```python
#finding "best-fit" parameter combination, plotting the best ones
Om0=0.3
OmL=0.5
H0=70
M=-19

z_grid = np.linspace(0.1, 2, 100)

dl = []
m_predicted = []
for z_value in z_grid:
    d_L = d_l(z_value, H0, Om0, OmL, atol=1e-16, rtol=1.e-8)
    m_pred = M + 5*np.log10(d_L) + 25
    dl.append(d_L)
    m_predicted.append(m_pred)

plt.figure(figsize=(7, 7))
plt.scatter(zsn, msn, marker='.', color='indigo', s=20, label='Supernovae Magnitude Data')
plt.plot(z_grid, m_predicted, color='red', lw=1, label='Predicted Magnitude')
plt.xlabel('z (Redshift)', fontsize=14)
plt.ylabel('Apparent Magnitude', fontsize=14)
plt.text(1.7, 14.5, 'M = -19\nOm0=0.3\nOmL=0.3', fontsize=12)
plt.legend(loc='best', frameon=False)
plt.show()
```

Discussion: 
In plotting this graph, I found that the best-fit parameters were M = -19, Om0 = 0.3, and OmL = 0.5. This aligns relatively well with the standard model parameters, where Om0 = 0.3 and OmL = 0.7 (approximately). Additionally, the best-fit magnitude aligns very well with the real supernova data, suggesting that the predicted model for cosmological expansion is correct and the universe is actually expanding.
