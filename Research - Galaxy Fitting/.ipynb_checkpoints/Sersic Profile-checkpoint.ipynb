**Overview:**

Visualising SÃ©rsic profiles with varying n indexes (including Gaussian, Exponential, Devaucouleurs) by plotting radius vs. surface brightness.

```python
import numpy as np
import matplotlib.pyplot as plt

def sersic_plot(mag, Re, r, n):
    k = 2 * n - 0.324
    return mag * np.exp(-k * ((r/Re)**(1/n) - 1))

mag = 20
n_index = np.array([0.25, 0.5, 1, 4, 8])  
radius = np.linspace(0, 100, 1000)
Re = 8
surface_brightness = []

for n in n_index:
    brightness = sersic_plot(mag, Re, radius, n)
    surface_brightness.append(brightness)

plt.figure(figsize=(8, 8))
for i, n in enumerate(n_index):
    plt.plot(radius, surface_brightness[i], label=f"n = {n}")

plt.axis([1e-1, 100, 1e-2, 1e5])
# plt.xscale('log')
plt.yscale('log')
plt.xlabel('Radius')
plt.ylabel('Surface Brightness')
plt.grid(ls='dotted', color='lightgray')
plt.legend(loc='best', frameon=False)
plt.show()
```
