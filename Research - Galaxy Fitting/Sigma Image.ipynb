**Overview:**

Converting inverse variance (FITS) files to sigma image (FITS) files.

```python
from astropy.io import fits
import numpy as np
from matplotlib import pylab as plt

# Load the FITS file
fits_file = 'LEDA1722581-invvar-r.fits.fz'
hdul = fits.open(fits_file)

hdul.info()

# Access the image data (assuming it's in the primary HDU)
image_data = hdul[1].data

# Ensure the data is in floating-point format for precision
image_data_float = image_data.astype(np.float32)

# Take the square root of each pixel value
sqrt_image_data = np.sqrt(image_data_float)

# Optionally normalize or scale the data if required (e.g., for visualization)
sqrt_image_data_normalized = (sqrt_image_data - sqrt_image_data.min()) / (sqrt_image_data.max() - sqrt_image_data.min())

# Save the result to a new FITS file
new_hdul = fits.HDUList([fits.PrimaryHDU(data=sqrt_image_data_normalized)])
new_hdul.writeto('LEDA1722581_sigma.fits', overwrite=True)

# Close the original FITS file
hdul.close()

# Plot the original and square root-transformed images
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.imshow(image_data, cmap='gray', origin='lower')
plt.colorbar(label='Pixel Value')
plt.title('Original Image')

# Square Root Image
plt.subplot(1, 2, 2)
plt.imshow(sqrt_image_data, cmap='gray', origin='lower')
plt.colorbar(label='Pixel Value (sqrt)')
plt.title('Square Root Transformed Image')

# Show the plots
plt.tight_layout()
plt.show()
```
